# Object Oriented (OO) Plots
## Graphs of a math function basically 

### Plots for the following functions
### $f(x) = x$
### $f(x) = x^2$
### $f(x) = x^3$

````python

# linspace(start, stop, number_of_samples)
# Generates numbers with constant difference
# In this case the difference of any two consecutive numebers is 0.02020202

x = np.linspace(0, 2, 100)

fig, ax = plt.subplots()
ax.plot(x, x, label='linear')
````

````python
fig, ax = plt.subplots()
ax.plot(x, x**2, label='quadratic')
````

````python
fig, ax = plt.subplots()
ax.plot(x, x**3, label='cubic')
````

## Merging the above the plots into one and making it look nice

````python
fig, ax = plt.subplots()

ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')
ax.legend() # We have added labels to show in the legend, else it is not required
````
